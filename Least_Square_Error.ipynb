```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt 

#data input
x = np.array([2,3,5,7,11])
y = np.array([1,3,4,5,6])

#number of data points
n = len(x)

#calculate sum required to least square method
Summation_x = np.sum(x)
summation_y = np.sum(y)
summation_xy = np.sum(x * y)
summation_x_square = np.sum(x ** 2)

#calculate slope(m) and intercept(c)
m = ((n * summation_xy) - (Summation_x * summation_y)) / ((n * summation_x_square) - (Summation_x ** 2))

c = (summation_y - (m * Summation_x)) / n

#print calculate values of m and c 
print(f"slope(m) : {m}")
print(f"Intercept(c):{c}")

#plotting  the data points
plt.scatter(x,y,color = 'blue',label = 'Data points')

#plotting the Regression line
plt.plot(x,m * x + c,color = 'red',label = f'Regression line : y={m : .2f}x + {c : .2f}')

#labels the axes
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression using Least Squares Method')

#Adding legend
plt.legend()

# Display the plot
plt.show()
```
